```python
import geopandas
import os, re
import plotly.graph_objects as go
```

```python
ac_shp_fl = r"./TN election analysis/maps-master/assembly-constituencies/India_AC.shp"
pc_shp_fl = r"./TN election analysis/maps-master/parliamentary-constituencies/india_pc_2019.shp"
```

```python
ac_shp_df = geopandas.read_file(ac_shp_fl)
ac_shp_tn_df = ac_shp_df[ac_shp_df['ST_NAME'] =='TAMIL NADU']
ac_shp_tn_df.columns
```

```python
import plotly.express as px
import geopandas as gpd

df = px.data.election()
display(df.head())
geo_df = gpd.GeoDataFrame.from_features(
    px.data.election_geojson()["features"]
).merge(df, on="district").set_index("district")
geo_df
```

```python

fig = px.choropleth(geo_df,
                   geojson=geo_df.geometry,
                   locations=geo_df.index,
                   color="Joly",
                   projection="mercator")
fig.update_geos(fitbounds="locations", visible=False)
fig.show()
```
